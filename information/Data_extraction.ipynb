```python
#Importing necessary modules and libraries
#Data processing
import pandas as pd
import random
import json
```

```python
#Reading all of the data from the dataset providen
data = []
for i in range(0, 12, 1) :
    df_sample = pd.read_parquet(path = f'/content/drive/MyDrive/data/train_data_{i}.pq')
    data.append(df_sample)

df = pd.concat(data)
df.head()
```

```python
#Reading and outputting dataframe with target variable
target_df = pd.read_csv(filepath_or_buffer = '/content/drive/MyDrive/data/train_target.csv')
target_df.head()
```

```python
#Extraction 5 sample values for class 1 and class 0
flag_1_ids = target_df[target_df['flag'] == 1].sample(n = 5)['id'].tolist()
```

```python
flag_0_ids = target_df[target_df['flag'] == 0].sample(n = 5)['id'].tolist()
```

```python
for i in flag_1_ids + flag_0_ids :
    temp_df = df[df['id'] == i]
    temp_dict = temp_df.to_dict()
    filename = f"dict_{i}.txt"

    with open(filename, 'w') as file:
        json.dump(temp_dict, file, indent=4)
    print(f"Dictionary saved to {filename}")
```
